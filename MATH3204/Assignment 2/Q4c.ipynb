```python
import numpy as np
from scipy import sparse
from scipy.sparse import linalg
import matplotlib.pyplot as plt
import time
```

```python
num = [100,200,500,1000,2000,5000,10000]
time1 = [] 
time2 = []

for n in num:
    A = sparse.random(n,n,0.1,format='csc')
    b = np.ones((n,1))
    # time taken using inv
    start1 = time.time()
    x = sparse.linalg.inv(A)*b
    end1 = time.time()
    time1.append(end1-start1)
    # time taken using backslash
    start2 = time.time()
    x = sparse.linalg.spsolve(A,b)
    end2 = time.time()
    time2.append(end2-start2)

plt.xlabel('n')
plt.ylabel('time')
plt.plot(num,time1, 'r',label='inv')
plt.plot(num,time2, 'b',label='backslash')
plt.legend(loc='best')
```
