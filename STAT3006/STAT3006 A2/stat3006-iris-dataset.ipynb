```r
# Variables for the fitted Setosa data
xA = data.matrix(fitted_setosa[1:2]) # Sepal Length & Width in Setosa
xB = data.matrix(fitted_setosa[3:4]) # Petal Length & Width in Setosa

fitted_setosa_mu = colMeans(fitted_setosa)
muA = data.matrix(fitted_setosa_mu[1:2]) # mean of xA
muB = data.matrix(fitted_setosa_mu[3:4]) # mean of xB

fitted_versicolor_sig = cov(fitted_setosa)
sigAA = data.matrix(fitted_setosa_sig[1:2,1:2])
sigBB = data.matrix(fitted_setosa_sig[3:4,3:4])
sigAB = data.matrix(fitted_setosa_sig[1:2,3:4])
sigBA = data.matrix(fitted_setosa_sig[3:4,1:2])

# Conditional distribution for Petal Length/Width
con_muA = muA + sigAB %*% solve(sigBB) %*% (colMeans(xB) - muB)
con_sigA = sigAA - sigAB %*% solve(sigBB) %*% sigBA

# Contour the conditional distribution
x.points <- seq(-1,1,length.out=100)
y.points <- x.points
z <- matrix(0,nrow=100,ncol=100)
mu <- c(1,1)
sigma <- matrix(c(2,1,1,1),nrow=2)
for (i in 1:100) {
    for (j in 1:100) {
        z[i,j] <- dmvnorm(c(x.points[i],y.points[j]),mean=con_muA,sigma=con_sigA)}}
plot(xB,xlim=c(-1,1),ylim=c(-1,1)) # Add points for Q7
contour(x.points,y.points,z)
```

# Part 2: Hypothesis testing

**Exericise 1:**

```r
compare = iris[Species!="setosa",1:5]
m1 <- manova(cbind(Sepal.Length,Sepal.Width,Petal.Length,Petal.Width) ~ Species, compare)
summary(m1, test = "Hotelling-Lawley")
```

**Exericise 2:**

```r
# Create empty vectors for sample size and p values
p_values <- vector()
sample_size <- vector()

for (num in c(10:100)){
    sample = sample_n(compare, num) # choose random subset
    m2 <- manova(cbind(Sepal.Length,Sepal.Width,Petal.Length,Petal.Width) ~ Species, sample)
    test <- summary(m2, test = "Hotelling-Lawley")
    p_values <- c(p_values,test$stats[1,6])
    sample_size <- c(sample_size,num)
    }

ggplot(data.frame(x=sample_size, y=p_values), aes(x=sample_size, y=p_values)) + geom_line() 
```

```r
library(cluster)

# Apply Gap Statistic for K-means for Iris
iris_gap <- clusGap(iris[-5], FUN = kmeans ,nstart = 20, K.max = 5, B = 100)
plot(iris_gap, main="Gap Statistic for Iris")

# Apply K-means clustering for Iris
iris_k2 <- kmeans(iris[-5], centers=2)
iris_k3 <- kmeans(iris[-5], centers=3)
iris_k4 <- kmeans(iris[-5], centers=4)
iris_k5 <- kmeans(iris[-5], centers=5)

# Plot K-means clustering for Iris
par(mfrow=c(2,2),oma = c(0, 0, 2, 0))
plot(iris[,c("Sepal.Length","Sepal.Width")], col = iris_k2$cluster, main = "k = 2")
plot(iris[,c("Sepal.Length","Sepal.Width")], col = iris_k3$cluster, main = "k = 3")
plot(iris[,c("Sepal.Length","Sepal.Width")], col = iris_k4$cluster, main = "k = 4")
plot(iris[,c("Sepal.Length","Sepal.Width")], col = iris_k5$cluster, main = "k = 5")
mtext("K-means Clustering for Iris", outer = TRUE, cex = 1.5,font=2)
```

```r
library(mclust)

# Apply BIC for mixture model for Iris
iris_BIC <- mclustBIC(iris[-5], G=seq(1,5), modelNames=c("VVV"))
plot(iris_BIC, main="BIC for Iris")

# Apply GMM clustering for Iris
iris_g2 <- Mclust(iris[-5], 2)
iris_g3 <- Mclust(iris[-5], 3)
iris_g4 <- Mclust(iris[-5], 4)
iris_g5 <- Mclust(iris[-5], 5)

# Plot GMM clustering for Iris
plot(iris_g2,what = "classification", main="k = 2")
plot(iris_g3,what = "classification", main="k = 3")
plot(iris_g4,what = "classification", main="k = 4")
plot(iris_g5,what = "classification", main="k = 5")
```

```r
# parameters for 3 components
props = art_g3$parameters$pro
means = art_g3$parameters$mean
sigma = art_g3$parameters$variance$sigma
cov1 <- matrix(sigma[1:4],2,2)
cov2 <- matrix(sigma[5:8],2,2)
cov3 <- matrix(sigma[9:12],2,2)

# Set range for contour plots
xs = seq(-5, 5, 0.1)
ys = seq(-6, 5, 0.1)
valsx = length(xs)
valsy = length(ys)
z <- array(0, c(1,valsx,valsy))
plot(art, xlim=c(-5,5), ylim=c(-6,5), pch=20, cex=0.1, main = "fitted mixture distribution with 3 components for Artificial")

# Contour plot of component 1
mu1 = means[,1]
issig1 = solve(cov1)
const1 = props[1]/((2*pi)^(2/2)*sqrt(det(cov1)))
for (i in 0:(valsx-1)){
    for (j in 0:(valsy-1)){
        x = -5 + 0.1*i
        y = -6 + 0.1*j
        xy = c(x,y)
        z[1, (i+1), (j+1)] = const1*exp(-0.5*(t(xy) - t(mu1)) %*% issig1 %*% (xy-mu1))
    }
}
contour(xs,ys,z[1,,], add=TRUE, drawlabels=T, col=2, levels=c(0.01,0.001))

# Contour plot of component 2
mu2 = means[,2]
issig2 = solve(cov2)
const2 = props[2]/((2*pi)^(2/2)*sqrt(det(cov2)))
for (i in 0:(valsx-1)){
    for (j in 0:(valsy-1)){
        x = -5 + 0.1*i
        y = -6 + 0.1*j
        xy = c(x,y)
        z[1, (i+1), (j+1)] = const2*exp(-0.5*(t(xy) - t(mu2)) %*% issig2 %*% (xy-mu2))
    }
}
contour(xs,ys,z[1,,], add=TRUE, drawlabels=T, col=3, levels=c(0.01,0.001))

# Contour plot of component 3
mu3 = means[,3]
issig3 = solve(cov3)
const3 = props[3]/((2*pi)^(2/2)*sqrt(det(cov3)))
for (i in 0:(valsx-1)){
    for (j in 0:(valsy-1)){
        x = -5 + 0.1*i
        y = -6 + 0.1*j
        xy = c(x,y)
        z[1, (i+1), (j+1)] = const3*exp(-0.5*(t(xy) - t(mu3)) %*% issig3 %*% (xy-mu3))
    }
}
contour(xs,ys,z[1,,], add=TRUE, drawlabels=T, col=4, levels=c(0.01,0.001))
```
