### Cylinder: non-vanishing vector field

```python
E3.<x,y,z> = EuclideanSpace()
cart = E3.default_chart()
Cyl = Manifold(2, 'C', start_index=1)
CC.<eta,v> = Cyl.chart(r"eta:\eta:(0,2*pi):periodic v")
Phi = Cyl.diff_map(E3, {(CC, cart):(cos(eta),sin(eta),v)})
W = Cyl.vector_field((1,0))
W.display()
(CC.plot(cart, mapping=Phi, number_values=8, ranges={v:(-1,1)})
 + W.plot(cart, mapping=Phi, number_values=8, ranges={v:(-1,1)}, width=1/3, color='green'))
```

```python
Phi.display()
```

### Torus: non-vanishing vector field

```python
E3.<x,y,z> = EuclideanSpace()
cart = E3.cartesian_coordinates()
T2 = Manifold(2, 'T2', latex_name=r"\mathbb{T}^2", start_index=1, ambient=E3, structure="Riemannian")
CT2.<u,v> = T2.chart('u:(0,2*pi):periodic v:(0,2*pi):periodic')
FT2 = CT2.frame()
Psi = T2.diff_map(E3, {(CT2, cart):
                        [3*(2+cos(v))*cos(u),
                         3*(2+cos(v))*sin(u),
                         3*(2+sin(v))]},
                   name='Psi', latex_name=r"\Psi")
T2.set_immersion(Psi)
F = T2.vector_field((1/2,0))

nvals = 6
(CT2.plot(cart, mapping=Psi, number_values=nvals)
 + F.plot(cart, mapping=Psi, number_values=nvals, color='blue'))
```

```python
Psi.display()
```

### Sphere

```python
E3.<x,y,z> = EuclideanSpace()
cart = E3.cartesian_coordinates()
T2 = Manifold(2, 'T2', latex_name=r"\mathbb{T}^2", start_index=1, ambient=E3, structure="Riemannian")
CT2.<u,v> = T2.chart('u:(0,2*pi):periodic v:(0,2*pi):periodic')
FT2 = CT2.frame()
Psi = T2.diff_map(E3, {(CT2, cart):
                        [cos(v)*cos(u),
                        cos(v)*sin(u),
                         sin(v)]},
                   name='Psi', latex_name=r"\Psi")
T2.set_immersion(Psi)
F = T2.vector_field((1,0))

nvals = 6
(CT2.plot(cart, mapping=Psi, number_values=nvals)
 + F.plot(cart, mapping=Psi, number_values=nvals, color='blue',arrowsize=0.5))
```

```python
Psi.display()
```
